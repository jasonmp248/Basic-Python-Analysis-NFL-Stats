```python
import pandas as pd
import nfl_data_py as nfl
```

```python
pbp_2021 = nfl.import_pbp_data([2023])
pd.set_option('display.max_columns', None)
```

```python
pbp_passer = pbp_2021.query('play_type == "pass" & air_yards.notnull()').groupby(['passer_id', 'passer']).agg({"air_yards": ["count", "mean"]})     #air_yards.notnull() -> removes missing data
pbp_passer
```

```python
##flattening muklti-index columns -> seems multi-index cols happen when you aggregate based on field

##"The pandas package also requires reformatting the column heads via a list() function and changing the header from being two rows to a single row via map()."

#pbp_passer.columns.values    #view multi-index cols; wanna join together
pbp_passer.columns = list(map("_".join, pbp_passer.columns.values))
pbp_passer
```

```python
pbp_passer.query("air_yards_count > 100").sort_values(by="air_yards_mean", ascending = False)
```
